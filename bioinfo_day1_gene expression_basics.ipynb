```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Pandas version:", pd.__version__)
print("NumPy version:", np.__version__)
```

```python
genes = ["TGFB", "IL4", "IFNG", "IL1B", "TNFA"]
samples = ["Tumor1","Tumor2","Normal1","Normal2"]

data = np.random.rand(len(genes), len(samples))*100

df = pd.DataFrame(data, index=genes, columns=samples)
df
```

```python
#View first few rows of the DataFrame
df.head
```

```python
#Transpose the DataFrame
df.T
```

```python
#Select a specific gene (row)'
df.loc["TGFB"]
```

```python
#Select specific samples (columns)
df[["Tumor1", "Tumor2"]]
```

```python
df>50 
#Boolean filter to see values above 50
```

```python
#Compute mean expression per gene
df.mean(axis=1)
```

```python
#Compute mean expression per sample
df.mean(axis=0)
```

```python
#Apply log2 transformation
log_df = np.log2(df + 1)
```

```python
#check log2 transformation
log_df.head(5)
```

```python
#Summary stats
df.describe()
```

```python
#Compute mean per gene using apply
df.apply(np.mean,axis=1)
```

```python
#Plot histogram of all values with alpha=transparency and bins= number of bars
df.plot.hist(alpha=0.5,bins=10)
```
